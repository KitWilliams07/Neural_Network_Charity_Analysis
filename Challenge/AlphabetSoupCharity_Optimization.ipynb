```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python
remove_names_df = application_df.drop(columns=['EIN','NAME'],axis=1)
remove_names_df
```

```python
# Cleaning Application Types
```

```python
remove_name_counts = remove_names_df.APPLICATION_TYPE.value_counts()
remove_name_counts
```

```python
app_type = remove_name_counts[remove_name_counts < 1000].index

for i in app_type:
    remove_names_df.APPLICATION_TYPE = remove_names_df.APPLICATION_TYPE.replace(i,'Other')
```

```python
# Cleaning Affiliation 
```

```python
remove_names_df.AFFILIATION.value_counts()
```

```python
# CLeaning Classification
```

```python
class_counts = remove_names_df.CLASSIFICATION.value_counts()
class_counts
```

```python
class_remove = class_counts[class_counts < 800].index

for item in class_remove:
    remove_names_df.CLASSIFICATION = remove_names_df.CLASSIFICATION.replace(item,'Other')
```

```python
# Cleaning Ask Amount
remove_names_df.ASK_AMT.value_counts()
```

```python
# This is incredibly messy, will drop this column for now and re-evaluate accuracy.
remove_names_df = remove_names_df.drop(columns=['ASK_AMT'])
remove_names_df
```

```python
# One Hot Encoder
categorical_vars = ['APPLICATION_TYPE','AFFILIATION','CLASSIFICATION','USE_CASE','ORGANIZATION','INCOME_AMT','SPECIAL_CONSIDERATIONS']


# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
array_encoded = enc.fit_transform(remove_names_df[categorical_vars])
encode_df = pd.DataFrame(array_encoded)

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(categorical_vars)
encode_df.head()
```

```python
merged_df = pd.concat([remove_names_df,encode_df],axis=1)
merged_df
```

```python
merged_df = merged_df.drop(columns=['APPLICATION_TYPE','AFFILIATION','CLASSIFICATION','USE_CASE','ORGANIZATION','SPECIAL_CONSIDERATIONS','INCOME_AMT'],axis=1)
```

```python
merged_df
```

```python
# Split the data 
y = merged_df.IS_SUCCESSFUL.values
X = merged_df.drop(columns=['IS_SUCCESSFUL'],axis=1).values

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
# Scale the data
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
num_input = len(X_train[0])
num_input
```

```python
# Create Neural Network
num_input = len(X_train[0])

nn = tf.keras.models.Sequential()

# First Hidden Layer
nn.add(tf.keras.layers.Dense(units=100,input_dim=num_input,activation='relu'))

# Second Hidden Layer
nn.add(tf.keras.layers.Dense(units=50,activation='relu'))

# Third Hidden Layer
nn.add(tf.keras.layers.Dense(units=10,activation='relu'))

# Output Layer
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
import os
from keras.callbacks import ModelCheckpoint

os.makedirs("optimized_checkpoints/",exist_ok=True)
checkpoint_path = "optimized_checkpoints/weights.{epoch:02d}.hdf5"

cp_callback = ModelCheckpoint(filepath=checkpoint_path,verbose=1,save_weights_only=True,save_freq='epoch',period=5)
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=200,callbacks=[cp_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
